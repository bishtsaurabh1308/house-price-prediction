```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('real_estate_regression.xlsx')
df.head(2)
```

```python
df = df.drop(columns=['Property_ID'])
df.head(2)
```

```python
nums = df.select_dtypes(include='number')
nums
```

```python
for col in nums:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

```python
nums.skew()
```

```python
df.duplicated().sum()
```

```python
cols = df.select_dtypes(include='object')
cols
```

```python
df.dropna(subset=['Property_Price'],inplace=True)
```

```python
col_nums = df.select_dtypes(include='number').columns

for col in col_nums:
    df[col] = df[col].fillna(df[col].mean())
```

```python
cat_cols = df.select_dtypes(include='object').columns

for i in cat_cols:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df.columns
```

```python
change_dtype = ['Bedrooms','Bathrooms','Year_Built','Nearby_Schools','Transport_Accessibility','Green_Space_Availability','Local_Amenities']
```

```python
df[change_dtype] = df[change_dtype].round().astype(int)
df[change_dtype].head(3)
```

```python
round_columns = ['Size_sqft','Property_Tax','Crime_Rate','HOA_Fees','Average_Income_Area','Property_Price']
df[round_columns] = df[round_columns].round(2)
df.sample(10)
```

```python
df.to_csv('house_price_cleaned.csv', index=False)
```
